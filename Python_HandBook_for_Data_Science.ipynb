# Attributes of arrays: Determining the size, shape, memory consumption, and data types of arrays
# Indexing of arrays: Getting and setting the value of individual array elements
# Slicing of arrays: Getting and setting smaller subarrays within a larger array
# Reshaping of arrays: Changing the shape of a given array
# Joining and splitting of arrays: Combining multiple arrays into one, and splitting one array into many

```python

import numpy as np
np.random.seed(0)
x1 = np.random.randint(10, size=6)
x2 = np.random.randint(10, size=(3,4))
x3 = np.random.randint(10, size=(3,4,5))
```

```python
# attributes: dimension, shape, size
print('dimension', x3.ndim)
print('shape', x3.shape)
print('size', x3.size)
```

```python
# Array Slicing
# array[start:stop:step]
x1
print(x1)
x1[1::1]
```

```python
# Slicing multi dimensional Array
print(x2)
# arr[1:2:1,1:2:1]. first slice is for rows and second slice is for columns
x2[:2,:3] # This gives first two rows and all columns
x2[::,::2] # this gives all rows with columns one after other
x2[::-1, ::-1] # this gives reverse of elements in array
```

```python
# sub arrays are views instead of copies
print(x2)
x2_sub = x2[:2,:2]
print(x2_sub)
x2_sub[0,0] = 99
print(x2_sub)
print(x2)
print('------------')

# use copy method such that if we change anything in sub array it wont reflect original array
x2_sub_copy = x2[:2,:2].copy()
print(x2_sub_copy)
x2_sub_copy[0,0] = 42
print(x2_sub_copy)
print(x2)
```

# Concatenation of Arrays
## concatenate, vstack, hstack


```python
x = np.random.randint(1,10,3)
y = np.random.randint(1,10,3)
```

```python
z = np.concatenate([x,y]).reshape((2,3))
print(z)
```

```python
# vstack is vertical stack and hstack is horizonta stack
x = np.array([[1,2],[2,3]])
print(x)
y = np.vstack([x,[2,5]])
print(y)
z = np.hstack([x,[[99],[99]]])
print(z)
```

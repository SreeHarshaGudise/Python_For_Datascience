# Pandas Manipulations

```python
# Pandas Series Object
import pandas as pd
x = pd.Series([1,2,3,4,5])
print(x)
# As it contains the values and indices as attributes
print(x.values)
print(x.index)

# we can use non contiguos indices as well
y = pd.Series(data = [3.,4,5,1], index=[2,5,1,7])
print(y)
```

```python
# we can convert python dictionary to pandas series

x = pd.Series({2:'a',5:'c',3:'d'})
x
```

```python
# Data Frame: Its a two dimensional series object as its a generilization of numpy array

population_dict = {'California': 38332521,
                   'Texas': 26448193,
                   'New York': 19651127,
                   'Florida': 19552860,
                   'Illinois': 12882135}
population = pd.Series(population_dict)
population
area_dict = {'California': 423967, 'Texas': 695662, 'New York': 141297,
             'Florida': 170312, 'Illinois': 149995}
area = pd.Series(area_dict)

states = pd.DataFrame({'area':area,'population': population})
states['area']
```

```python
# Pandas Index Object which is immutable array

p = pd.Index(data=[1,2,3,4,5])
p
p[0] = 3
```

```python
# attributes of index object
p.size, p.shape, p.ndim
```
